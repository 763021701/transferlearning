```python
import numpy as np
import matplotlib.pyplot as plt

# 假设我们有以下数据
# 真实数据的特征函数值
phi_real = np.array([0.8 + 0.6j, 0.5 + 0.3j, 0.2 + 0.4j])
# 合成数据的特征函数值
phi_synthetic = np.array([0.75 + 0.65j, 0.45 + 0.35j, 0.15 + 0.45j])

# 提取幅度和相位
amplitude_real = np.abs(phi_real)
phase_real = np.angle(phi_real)

amplitude_synthetic = np.abs(phi_synthetic)
phase_synthetic = np.angle(phi_synthetic)

# 绘制复平面图
plt.figure(figsize=(10, 5))

# 真实数据
plt.subplot(1, 2, 1)
plt.title("Real Data")
# 为每个向量指定起点 (0, 0)
plt.quiver([0] * len(phi_real), [0] * len(phi_real), 
           amplitude_real * np.cos(phase_real), amplitude_real * np.sin(phase_real),
           angles='xy', scale_units='xy', scale=1, color='blue', label='Real')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.legend()

# 合成数据
plt.subplot(1, 2, 2)
plt.title("Synthetic Data")
# 为每个向量指定起点 (0, 0)
plt.quiver([0] * len(phi_synthetic), [0] * len(phi_synthetic), 
           amplitude_synthetic * np.cos(phase_synthetic), amplitude_synthetic * np.sin(phase_synthetic),
           angles='xy', scale_units='xy', scale=1, color='red', label='Synthetic')
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 假设我们有以下数据
# 真实数据的特征函数值
phi_real = np.array([0.8 + 0.6j, 0.5 + 0.3j, 0.2 + 0.4j])
# 合成数据的特征函数值
phi_synthetic = np.array([0.75 + 0.65j, 0.45 + 0.35j, 0.15 + 0.45j])

# 提取幅度和相位
amplitude_real = np.abs(phi_real)
phase_real = np.angle(phi_real)

amplitude_synthetic = np.abs(phi_synthetic)
phase_synthetic = np.angle(phi_synthetic)

# 绘制复平面图
plt.figure(figsize=(10, 5))

# 真实数据
plt.subplot(1, 2, 1)
plt.title("Real Data")
for i in range(len(phi_real)):
    # 绘制小扇形
    theta = np.linspace(phase_real[i] - 0.1, phase_real[i] + 0.1, 100)
    r = amplitude_real[i]
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.fill_between(x, y, color='blue', alpha=0.3)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.legend(['Real'])

# 合成数据
plt.subplot(1, 2, 2)
plt.title("Synthetic Data")
for i in range(len(phi_synthetic)):
    # 绘制小扇形
    theta = np.linspace(phase_synthetic[i] - 0.1, phase_synthetic[i] + 0.1, 100)
    r = amplitude_synthetic[i]
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.fill_between(x, y, color='red', alpha=0.3)
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.xlabel("Real")
plt.ylabel("Imaginary")
plt.legend(['Synthetic'])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge, Circle

# 假设我们有以下数据
# 真实数据的特征函数值
phi_real = np.array([0.8 + 0.6j, 0.5 + 0.3j, 0.5 + 0j])

# 提取幅度和相位
amplitude_real = np.abs(phi_real)
phase_real = np.angle(phi_real)

print(amplitude_real)
print(phase_real)

# 绘制复平面图
plt.figure(figsize=(10, 5))

# 真实数据
plt.subplot(1, 2, 1)
plt.title("Real Data")
ax = plt.gca()  # 获取当前子图的轴

# 绘制圆形边界
circle = Circle((0, 0), radius=1, fill=False, color='black', linestyle='--')
ax.add_patch(circle)

colors = ['red', 'blue', 'yellow']

# 绘制小扇形
for i in range(len(phi_real)):
    wedge = Wedge((0, 0), amplitude_real[i], np.degrees(phase_real[i] - 0.1), np.degrees(phase_real[i] + 0.1), color=colors[i], alpha=0.3)
    ax.add_patch(wedge)

# 设置坐标轴范围为圆形
ax.set_xlim(-1, 1)
ax.set_ylim(-1, 1)
ax.set_aspect('equal', adjustable='box')  # 保持比例
ax.set_xticks([])  # 隐藏x轴刻度
ax.set_yticks([])  # 隐藏y轴刻度
ax.spines['left'].set_visible(False)  # 隐藏左边框
ax.spines['right'].set_visible(False)  # 隐藏右边框
ax.spines['top'].set_visible(False)  # 隐藏上边框
ax.spines['bottom'].set_visible(False)  # 隐藏下边框

plt.tight_layout()
plt.show()
```
